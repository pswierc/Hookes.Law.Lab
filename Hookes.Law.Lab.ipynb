```python
#method 2

import physplot as phys
import numpy as np


m1 = np.array([50,100,150,200,250,300])/1000
s1 = (np.array([4.07,5.7,7,8.10,9.06,9.87]) / 10 ) ** 2
errm = np.array([1,1,1,1,1,1])/1000

m2 = np.array([150,170,190,210,250,300])/1000
s2 = (np.array([13.9,14.95,15.83,16.60,18.18,19.86]) / 10 ) ** 2
errm2 = np.array([1,1,1,1,1,1])/1000

phys.plotfit(s1,m1,errm,'Small k Spring','T^2 (s^2)','m*kg','kg/s^2')
phys.plotfit(s2,m2,errm2,'Large k Spring','T^2 (s^2)','m*kg','kg/s^2')
```
